```python
def split_file(file_path, split_size,output_dir):
    chunk_size = split_size  # 设置每个块的大小
    file_number = 1
    for chunk in pd.read_csv(file_path, sep='\t',chunksize=chunk_size, error_bad_lines=False,low_memory=False):
        columns_to_keep = [
        "kingdom", "phylum", "class", "order", "family", "genus", "species",
        "stateProvince", "decimalLatitude", "decimalLongitude", "day", "month", 
        "year", "taxonKey", "speciesKey", "identifiedBy"]
        chunk= chunk[columns_to_keep]
        chunk.to_csv(f'{output_dir}iNaturalist{file_number}.txt', index=False)  # 将每个块保存为独立的文本文件
        file_number += 1
```

```python
# Path to the ZIP file
zip_path = '0022035-241024112534372.zip'

# Output directory for year-based CSV files
output_dir = 'yearly_csvs'
os.makedirs(output_dir, exist_ok=True)  # Create the directory if it doesn't exist

# Open the ZIP file and read the CSV with tab as the delimiter
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    csv_filename = '0022035-241024112534372.csv'
    
    # Check if the specified CSV is in the ZIP file
    if csv_filename in zip_ref.namelist():
        # Load the CSV into a DataFrame using tab as the delimiter
        with zip_ref.open(csv_filename) as file:
            df = pd.read_csv(file, delimiter='\t', on_bad_lines='skip')
        
        # Check if 'dateidentified' column exists
        if 'dateidentified' in df.columns:
            # Extract the year from 'dateidentified' column
            df['year'] = df['dateidentified'].astype(str).str[:4]
            
            # Split the DataFrame by year and save each as a separate CSV
            for year in df['year'].unique():
                year_df = df[df['year'] == year]
                output_path = os.path.join(output_dir, f"{year}_data.csv")
                year_df.to_csv(output_path, index=False)
                print(f"Saved CSV for year {year} at: {output_path}")
        else:
            print("Column 'dateidentified' not found in the CSV file.")
    else:
        print(f"File '{csv_filename}' not found in the ZIP archive.")

```
